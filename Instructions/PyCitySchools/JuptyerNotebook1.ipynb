# PyCity Schools Analysis

```python
#Jupt# Dependencies and Setup
import pandas as pd

#File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

#Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

# District Summary

```python
#Calculating the total number of schools and students 
school_total = len(school_data_complete["school_name"].unique())
student_total = school_data_complete["student_name"].count()
#school_total 
#student_total 

#Calculating the total budget 
total_budget = school_data["budget"].sum()
#total_budget

#Calulating now the average math score 
aver_math_score = school_data_complete["math_score"].mean()
#aver_math_score

#Calculating the average reading score 
aver_read_score = school_data_complete["reading_score"].mean()
#aver_read_score

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
ovr_pass_rate = (aver_math_score + aver_read_score)/2
#ovr_pass_rate
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
passing_math_tot = school_data_complete[(school_data_complete["math_score"]>=70)].count()["student_name"]

#percent math passing
percent_math_pass = passing_math_tot / float(student_total) *100
#percent_math_pass

#Calculate the percentage of students with a passing reading score (70 or greater)
passing_read_tot = school_data_complete[(school_data_complete["reading_score"]>=70)].count()["student_name"]
#percent_read_pass
percent_read_pass = passing_read_tot / float(student_total) *100


#Calculate % overall passing (the percentage of students who passed math AND reading)
percent_ovr_pass = (percent_math_pass + percent_read_pass)/2
#Print percent_ovr_pass
```

```python
#Create a dataframe to hold the above results in district summary 
district_summary =[{"Total Schools": (school_total),
                    "Total Students": (student_total),
                    "Total Budget": (total_budget),
                    "Average Math Score": (aver_math_score),
                    "Average Reading Score": (aver_read_score),
                    "% Passing Math": (percent_math_pass),
                    "% Passing Reading": (percent_read_pass),
                    "% Overall Passing": (percent_ovr_pass)}]
district_summary_display = pd.DataFrame(district_summary)

#Print or Display
district_summary_display
```

```python
# Additional Formatting and data cleaning using map functions
district_summary_display["Total Students"]=district_summary_display["Total Students"].map("{:,.2f}%".format)
district_summary_display["Total Budget"]=district_summary_display["Total Budget"].map("{:,.2f}%".format)
district_summary_display["% Passing Math"]=district_summary_display["% Passing Math"].map("{:,.2f}%".format)
district_summary_display["% Passing Reading"]=district_summary_display["% Passing Reading"].map("{:,.2f}%".format)
district_summary_display["% Overall Passing"]=district_summary_display["% Overall Passing"].map("{:,.2f}%".format)

#Print or Display
district_summary_display
```

# School Summary 

```python
#Create an overview table that summarizes key metrics about each school, including:
#Determine School Type
school_types = school_data.set_index(["school_name"])["type"]

#Calculate total amount of students
school_count = school_data_complete["school_name"].value_counts()

#Calculate the total school budget 
per_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
budget_stu_tot = per_budget / per_school_count

#Calculate the average test scores of school 
per_math_ss = school_data_complete.groupby(["school_name"]).mean()["math_score"]
per_read_ss = school_data_complete.groupby(["school_name"]).mean()["reading_score"]

school_math_passed = school_data_complete[(school_data_complete["math_score"] >= 70)]
school_read_passed = school_data_complete [(school_data_complete["reading_score"] >= 70)]

passing_math = school_math_passed.groupby(["school_name"]).count()["student_name"] / school_count * 100
passing_read = school_read_passed.groupby(["school_name"]).count()["student_name"] / school_count * 100
ovr_passing = (passing_math + passing_read) /2

#Print or Display                                    
school_summary = pd.DataFrame({"School Type": school_types,
                               "Total Students": school_count,
                               "Total School Budget": per_budget,
                               "Per Student Budget": budget_stu_tot, 
                               "Average Math Score": per_math_ss,
                               "Average Reading Score": per_read_ss,
                               "% Passing Math": passing_math,
                               "% Passing Reading": passing_read,
                               "% Overall Passing Rate": ovr_passing,})

                           
```

```python
# Additional Formatting and data cleaning using map functions
school_summary = school_summary[["School Type", "Total Students", "Total School Budget", "Per Student Budget",
                                 "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading",
                                 "% Overall Passing Rate"]]

school_summary["Per Student Budget"]=school_summary["Per Student Budget"].map("${:,.2f}".format)
school_summary["Total School Budget"]=school_summary["Total School Budget"].map("${:,.2f}".format)

#Print or Display
school_summary
```

# Scores by School Spending

```python
#Creating Bins 
spending_bins = [0, 585, 615, 645, 675]
amounts = ["<$585", "$585-615", "$615-645", "$645-675"]

#Spending based on the bins created
school_summary["Spending Ranges (Per Student)"] = pd.cut(budget_stu_tot, spending_bins, labels=amounts)

#Scores 
spend_math = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spend_read = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spend_math_pass = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spend_read_pass = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passer = (spend_math + spend_read) / 2

#Creating the data frame based off bins 
school_spending= pd.DataFrame({"Average Math Score": spend_math,
                               "Average Reading Score": spend_read,
                               "% Passing Math": spend_math_pass,
                               "% Passing Reading": spend_read_pass,
                               "% Overall Passing Rate": overall_passer,})
#school_spending

school_spending_display= pd.DataFrame(school_spending)

#Data cleaning 
school_spending_display["Average Math Score"]=school_spending_display["Average Math Score"].map("{:,.2f}".format)
school_spending_display["Average Reading Score"]=school_spending_display["Average Reading Score"].map("{:,.2f}%".format)
school_spending_display["% Passing Math"]=school_spending_display["% Passing Math"].map("{:,.2f}%".format)
school_spending_display["% Passing Reading"]=school_spending_display["% Passing Reading"].map("{:,.2f}%".format)
school_spending_display["% Overall Passing Rate"]=school_spending_display["% Overall Passing Rate"].map("{:,.2f}%".format)

#Print or Display 
school_spending_display
```
